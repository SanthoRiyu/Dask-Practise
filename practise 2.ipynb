```python
def fun1(num):
    tot = 1
    for i in range(1,num):
        tot*= i
    return tot
def fun2(num):
    tot = 1
    for i in range(1,num):
        tot+= i
    return tot
def fun3(a,b):
    return a*b
# fun1(5000),fun2(5000)
fun3(fun1(90000),fun2(90000))
```

```python
out1 = client.submit(fun1,90000)
out2 = client.submit(fun2,90000)
out3 = client.submit(fun3,out1,out2)
progress(out3)
```

```python
out3.result()
```
